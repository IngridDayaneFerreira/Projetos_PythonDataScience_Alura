## Exploratory Analysis

#### Average value by property type

```python
# Calculate the average of all prices

df['Valor'].mean()
```

```python
# Grouping by property type

df.groupby('Tipo').mean(numeric_only=True)
```

```python
# Applying only to price

df.groupby('Tipo')['Valor'].mean()
```

```python
df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_preco_tipo.plot(kind='barh',figsize=(14,10),color='blue')
```

#### Removing Commercial Property

```python
# Identifying properties type

df.Tipo.unique()
```

```python
# Removing the Residential properties from the array

imoveis_comerciais = ['Conjunto Comercial/Sala', 'Prédio Inteiro', 'Loja/Salão', 'Galpão/Depósito/Armazém', 'Casa Comercial', 'Terreno Padrão', 'Box/Garagem',
       'Loja Shopping/ Ct Comercial', 'Chácara', 'Loteamento/Condomínio', 'Sítio', 'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
df.query('@imoveis_comerciais not in Tipo')
```

```python
# Removing the Commercial properties from the dataframe

df_im_res = df.query('@imoveis_comerciais not in Tipo')
```

```python
df_im_res.Tipo.unique()
```

```python
df_preco_tipo2 = df_im_res.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_preco_tipo2.plot(kind='barh',figsize=(8,5),color='blue')
```

#### Percentage by property type

```python
df_im_res.Tipo.unique()
```

```python
# Calculating the percentage with normalize

df_im_res.Tipo.value_counts(normalize=True)
```

```python
# Transforming into dataframe

df_im_res.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percent_tipo = df_im_res.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percent_tipo = df_im_res.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')

df_percent_tipo.plot(kind='bar',figsize=(10,8),color='grey',xlabel='Tipos',ylabel='Percentual')
```
